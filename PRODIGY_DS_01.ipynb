```python
#importing required packages 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('Behavioral_Risk_Factor_Data__Tobacco_Use__2011_to_present_.csv')  
```

```python
df.columns
```

```python
df.head()  
```

```python
df.tail()
```

```python
categorical_variable = df['Gender']  

# Calculating the frequency of each category
category_counts = categorical_variable.value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
categorical_variable = df['Age'] 

# Calculating the frequency of each category
category_counts = categorical_variable.value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='yellow', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.ylabel('Frequency')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
race_counts = df['Race'].value_counts()
race_counts.plot(kind='bar', color='lightpink', edgecolor='black')
plt.title('Race Distribution')
plt.xlabel('Race')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Data_Value'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Data Value Distribution')
plt.xlabel('Data Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
